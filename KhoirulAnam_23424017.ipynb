Soal 2 - Sampling Resize

```python
import cv2
import matplotlib.pyplot as plt

# membaca gambar
img = cv2.imread('gojo-hollow-purple.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# resize gambar
img512 = cv2.resize(img, (512,512))
img256 = cv2.resize(img, (256,256))
img128 = cv2.resize(img, (128,128))
img64 = cv2.resize(img, (64,64))

# menampilkan hasil
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img512)
plt.title("512 x 512")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(img256)
plt.title("256 x 256")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(img128)
plt.title("128 x 128")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(img64)
plt.title("64 x 64")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# baca gambar grayscale
img = cv2.imread('gojo-hollow-purple.jpg', 0)

# fungsi kuantisasi
def quantization(image, levels):
    step = 256 // levels
    quant = np.floor(image / step) * step
    return quant.astype(np.uint8)

# 2 bit = 4 level
img_2bit = quantization(img, 4)

# 4 bit = 16 level
img_4bit = quantization(img, 16)

# 8 bit = 256 level
img_8bit = quantization(img, 256)

# tampilkan hasil
plt.figure(figsize=(10,5))

plt.subplot(1,4,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(img_2bit, cmap='gray')
plt.title("2 bit (4 level)")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(img_4bit, cmap='gray')
plt.title("4 bit (16 level)")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(img_8bit, cmap='gray')
plt.title("8 bit (256 level)")
plt.axis("off")

plt.show()
```
